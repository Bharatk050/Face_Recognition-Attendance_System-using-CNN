```python
import numpy as np
import os
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers
from tensorflow.keras import Model
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from google.colab import drive
drive.mount('/content/Images')
```

```python
base_dir = "/content/Images/MyDrive/Images"
train_dir = os.path.join(base_dir, 'Train')
test_dir = os.path.join(base_dir, 'Test')

Bharat_train_face_dir = os.path.join(train_dir, 'Bharat')
Shyam_train_face_dir = os.path.join(train_dir, 'Shyam')
Kedop_train_face_dir = os.path.join(train_dir, 'Kedop')

Bharat_test_face_dir = os.path.join(test_dir, 'Bharat')
Shyam_test_face_dir = os.path.join(test_dir, 'Shyam')
Kedop_test_face_dir = os.path.join(test_dir, 'Kedop')
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,)

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=20,
    class_mode='binary')

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=20,
    class_mode='binary')
```

```python
img_input = layers.Input(shape=(150, 150, 3))

x = layers.Conv2D(16, 3, activation='relu')(img_input)
x = layers.MaxPooling2D(2)(x)

x = layers.Conv2D(32, 3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)

x = layers.Conv2D(64, 3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)

x = layers.Flatten()(x)
x = layers.Dense(512, activation='relu')(x)
output = layers.Dense(1, activation='sigmoid')(x)

model = Model(img_input, output)
model.compile(loss='binary_crossentropy',
            optimizer=RMSprop(learning_rate=0.001),
            metrics=['acc'])
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch=5,
    epochs=10,
    validation_data=test_generator,
    validation_steps=4,
    verbose=2)
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc)
plt.plot(epochs, val_acc)
plt.title('Training and validation accuracy')

plt.figure()

plt.plot(epochs, loss)
plt.plot(epochs, val_loss)
plt.title('Training and validation loss')
```

```python
def preprocess(file_path):
    byte_img = tf.io.read_file(file_path)
    img  = tf.io.decode_jpeg(byte_img)

    img = tf.image.resize(img, (150, 150))
    img=  img / 255.0

    return img
```

```python

def verify(model, detection_threshold, verification_threshold):
    # Build results array
    results = []
    for image in os.listdir(os.path.join('Images', 'Train')):
        input_img = preprocess(os.path.join('Images', 'Input_Image', 'input_image.jpg'))
        validation_img = preprocess(os.path.join('Images', 'Train', image))

        # Make Predictions
        result = model.predict(list(np.expand_dims([input_img, validation_img], axis=1)))
        results.append(result)

    # Detection Threshold: Metric above which a prediciton is considered positive
    detection = np.sum(np.array(results) > detection_threshold)

    # Verification Threshold: Proportion of positive predictions / total positive samples
    verification = detection / len(os.listdir(os.path.join('application_data', 'verification_images')))
    verified = verification > verification_threshold

    return results, verified
```

```python
results, verified = verify(model=model, detection_threshold=0.5, verification_threshold=0.5)
```
